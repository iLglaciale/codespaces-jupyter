```python
import numpy as np
import matplotlib.pyplot as plt
```

# DEFINIZIONE DI UNA FUNZIONE

```python
def f(x) : 
    return(x**2-2*x-3)
y=f(x)
```

verifichiamo se funziona

```python
print(f(0.5))
```

### definiamo un vettore che comprende i valori da -10 a 10 ogni 0.001, sono 2001 valori

```python
x=np.linspace(-10,10,2001)

print(x)
```

## calcoliamo la pendenza di $f(x)$ per ogni $x$ defindendo un $\Delta x$ chiamato dx

```python
dx=0.01
m=(f(x+dx)-f(x))/dx 
print(m)
```

## plottiamo su un grafico la funzione $f(x)$ e la sua pendenza $m(x)$

```python
griglia=np.linspace(-10,10,21)

plt.plot(x,y) #assi
plt.plot(x,m) #funzioni
plt.grid() #griglia
plt.gca().set_aspect("equal") #asse x e y stessa scala
plt.axis([-10,10,-10,10]) #da che valori per x e y vogliamo il grafico
plt.axhline(0, color='black', linewidth=1) #traccio una linea nera a x=0
plt.axvline(0, color='black', linewidth=1) #traccio una linea nera a y=0
plt.xticks(griglia) #creo una griglia per l'asse x per l'intervallo di valori della prima riga di codice
plt.yticks(griglia) #creo una griglia per l'asse y per l'intervallo di valori della prima riga di codice
plt.show()
```

### eseguiamo una regressione polinomiale su $m(x)$

```python
np.polyfit(x,m,1)
```

### quindi $m(x)=2x-1.99$
